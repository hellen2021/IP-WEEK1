**<h4>1.3 Select only the Sub counties from the Moroto district, order them alphabetically and show their production of sorghum.<h4>**

```python
#question 3
#select all subcounties located in Moroto district
#order the sucounties alphabetically
#also select their total sorghum production

%%sql
SELECT SUBCOUNTY_NAME, DISTRICT_NAME,S_Prod_Tot	FROM uganda
WHERE DISTRICT_NAME = 'MOROTO'
ORDER BY SUBCOUNTY_NAME ASC;
```

**<h4>1.4 Compute the total Maize production per District.<h4>**

```python
#question 4
#total maize production per district
#summing up the maize productions in every single district

%%sql
SELECT DISTRICT_NAME, SUM(M_Prod_Tot)
FROM uganda 
GROUP BY DISTRICT_NAME;
```
