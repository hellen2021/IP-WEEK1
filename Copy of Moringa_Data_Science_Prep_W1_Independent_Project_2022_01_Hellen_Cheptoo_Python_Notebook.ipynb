**<h4> 1.1 Take in two values from a user (both in pounds) then convert those values to kilograms <h4>**

```python
#take input values from the user in pounds
pound1 = float(input("Enter the first value in pounds:"))
pound2 = float(input("Enter the second value in pounds:"))

#convert to kilograms
# formula: kilograms = pounds * 0.453592

#convert the first value
kilogram1 = pound1 * 0.453592
print("The first pound value in kilograms =", kilogram1,"kg")

#convert the second value
kilogram2 = pound2 * 0.453592
print("The second pound value in kilograms =", kilogram2,"kg")
```

**<h4> 1.2 Perform the sum of the values.<h4>**

```python
sum = kilogram1 + kilogram2

print(sum)
```

**<h4> 1.3 Perform the average of the values.<h4>**

```python
# e.g average of 2 and 4 = (2+4)/2
average = (kilogram1 + kilogram2)/2

print("The average of the two values =", average)
```

**<h4> 1.4 Find the difference between both values.<h4>**

```python
#for this case, i want to find the difference between the two values such that the smaller value is subtracted from the bigger value
#since the values are input by the user, we do not know which is bigger than the other, therefore,

if kilogram1 > kilogram2:
  diff = kilogram1 - kilogram2
else:
  diff = kilogram2 - kilogram1

#print the difference between the two
print("The difference between the two values =", diff)  
```

**<h4> 1.5 Find the quotient when one value is divided by the other.<h4>**

```python
#use floor division
quotient = kilogram1 // kilogram2

print("The quotient of the two values =", quotient)
```

```python
#check for value1
if kilogram1 % 2 == 0:
  print("kilogram1 is even")
else:
  print("kilogram1 is odd")

#check for value2
if kilogram2 % 2 == 0:
  print("kilogram2 is even")
else:
  print("kilogram2 is odd")

```
